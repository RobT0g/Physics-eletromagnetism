```python
ponto_esquerda = 0
set_esquerda = 0

ponto_direita = 0
set_direita = 0

muda_saque = True

output = input("Sequência: ")
comando = []
for letra in output:
    comando.append(letra)
    
for index in comando:
    
    #VERIFICAÇÃO ANTECIPADA NO LOOP
    if set_esquerda >= 2:
        print(f"{set_esquerda} (winner) - {set_direita}")
        break
    elif set_direita >=2:
        print(f"{set_esquerda} - (winner) {set_direita}")
        break
    #COMANDOS
    if(index == "S"):
        if(muda_saque):
            ponto_esquerda += 1
        else:
            ponto_direita += 1
    elif(index == "R"):
        if(muda_saque):
            ponto_direita +=1
            muda_saque = False
        else:
            ponto_esquerda +=1
            muda_saque = True        
    elif(index == "Q"):
        
        if(muda_saque):
            print(f"{set_esquerda} ({ponto_esquerda}*) - {set_direita} ({ponto_direita})")
        else:
            print(f"{set_esquerda} ({ponto_esquerda}) - {set_direita} ({ponto_direita}*)")
    
    #VERIFICAÇÕES
    if (ponto_esquerda >= 5) and ((ponto_esquerda - ponto_direita) >= 2):
        set_esquerda += 1
        ponto_direita = 0
        ponto_esquerda = 0
        muda_saque = True
    elif (ponto_direita >= 5) and ((ponto_direita - ponto_esquerda) >= 2):
        set_direita += 1
        ponto_direita = 0
        ponto_esquerda = 0
        muda_saque = False
```
